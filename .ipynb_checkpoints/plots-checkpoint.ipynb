```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
from plotnine import ggplot, aes, ggtitle, geom_line, geom_point, geom_boxplot, xlab, ylab, facet_grid
import altair as alt
import ipywidgets as widgets
from datetime import datetime
import numpy as np

cp = sns.color_palette()
```

Simulate time series data w/random walk

```python
date_range = pd.date_range(start='1/1/23', end='1/1/24')

size = len(date_range)

ts = pd.DataFrame()

steps = np.random.standard_normal(size)
steps[0] = 0
random_walk = np.cumsum(steps)

cats = ['A', 'B', 'C', 'D']

for cat in cats:
    df = pd.DataFrame(date_range, columns=['time'])
    steps = np.random.standard_normal(size)
    steps[0] = 0
    random_walk = np.cumsum(steps)
    df['value'] = random_walk
    df['cat'] = cat
    ts = pd.concat([ts, df])

ts.head()
```

```python
ts.to_csv('random_walk_timeseries.csv')
```

Time Series Plotting

```python
# Matplotlib
plt.style.use('classic')
fig, ax = plt.subplots(figsize=(7.5, 5))

for cat in ts["cat"].unique():
    tmp = ts[ts["cat"] == cat]
    ax.plot(tmp.time, tmp.value, label=cat)

ax.set(xlabel='Date', ylabel='Value', title='Random Walk Timeseries')
ax.legend()

plt.show()
```

```python
# Pandas
ts_pivot = ts.pivot(index='time', columns='cat', values='value')
fig, ax = plt.subplots(figsize=(7.5, 5))

ts_pivot.plot(ax=ax)

ax.set(xlabel='Date', ylabel='Value', title='Random Walk Timeseries')
ax.legend(loc=2)

plt.show()
```

```python
# Seaborn
sns.set_theme()
g = sns.FacetGrid(ts, hue='cat', height=8, aspect=1.25)
g.map(plt.plot, 'time', 'value')
g.ax.set(xlabel='Date',
         ylabel='Value',
         title='Random Walk Timeseries')
g.add_legend()
g.fig.autofmt_xdate()
```

```python
#ggplot
g = ggplot(ts, aes(x='time', y='value', color='cat')) + \
geom_line(size=1) + \
xlab('Date') + \
ylab('Value') + \
ggtitle('Random Walk Timeseries')
g
```

```python
#altair

alt.Chart(ts).mark_line().encode(
    x=alt.X('time').title('Date'),
    y=alt.Y('value').title('Value'),
    color=alt.Color('cat').title('Category'),
    tooltip=['value']
).properties(title='Random Walk Timeseries', width=600, height=400) #.interactive()
```

Scatter Plot

```python
iris = pd.read_csv("Iris.csv")

iris["Species"] = iris["Species"].apply(lambda x: x.split("-")[1]).copy()

iris.head()
```

```python
#matplotlib
fig, ax = plt.subplots(1, 1, figsize=(7.5, 7.5))

def scatter(group):
    plt.plot(group["PetalLengthCm"], group["PetalWidthCm"], 'o', label=group.name)

iris.groupby('Species').apply(scatter)
ax.set(xlabel='Petal Length', ylabel='Petal Width', title='Petal Width vs Length by Species')
ax.legend(loc=2, numpoints=1)
```

```python
# Seaborn
sns.set_theme()
g = sns.FacetGrid(iris, hue='Species', height=7.5, aspect=1)
g.map(plt.scatter, 'PetalLengthCm', 'PetalWidthCm')
g.ax.set(xlabel='Petal Length',
         ylabel='Petal Width',
         title='Petal Length vs Width by Species')
g.add_legend()
```

```python
#ggplot
g = ggplot(iris, aes(x='PetalLengthCm', y='PetalWidthCm', color='Species')) + \
geom_point(size=2)
g
```

```python
#altair

alt.Chart(iris).mark_point(filled=True).encode(
    x=alt.X('PetalLengthCm').title('Petal Length'),
    y=alt.Y('PetalWidthCm').title('Petal Width'),
    color=alt.Color('Species').title('Species'),
    tooltip=['PetalLengthCm', 'PetalWidthCm']
).properties(title='Petal Width vs Length', width=600, height=400)
```

Faceted

```python
#matplotlib
fig, ax = plt.subplots(1, 3, figsize=(15, 5),
                      sharex=True, sharey=True)

for i, s in enumerate(iris.Species.unique()):
    tmp = iris[iris['Species'] == s]
    ax[i].scatter(tmp.PetalLengthCm, tmp.PetalWidthCm, color=cp[i])
    ax[i].set(xlabel='Petal Length', ylabel='Petal Width', title=s)

fig.tight_layout()
```

```python
# Seaborn
sns.set_theme()
g = sns.FacetGrid(iris, col='Species', hue='Species', height=5)
g.map(plt.scatter, 'PetalLengthCm', 'PetalWidthCm')
```

```python
# Altair

c = alt.Chart(iris).mark_point(filled=True).encode(
    x=alt.X('PetalLengthCm').title('Petal Length'),
    y=alt.Y('PetalWidthCm').title('Petal Width'),
    color=alt.Color('Species').title('Species'),
    column=alt.Column('Species', title="Petal Width vs Length by Species"),
    tooltip=['PetalLengthCm', 'PetalWidthCm']
).properties(width=300, height=300)
c
```
