```python
import pandas as pd
import matplotlib.pyplot as plt
from plotnine import (ggplot, 
                    aes, 
                    ggtitle, 
                    geom_bar, 
                    coord_flip, 
                    geom_text, 
                    geom_histogram,
                    geom_line,
                    geom_point,
                    geom_boxplot,
                    xlab,
                    ylab,
                    facet_grid,
                    facet_wrap,
                    geom_smooth,
                    geom_col,
                    geom_density,
                    after_stat,
                    xlim,
                    scale_x_datetime)
```

Bar chart
Histogram
Scatter plot (faceted by color, points sizing, all kinds of faceting)
Scatter with regression line + 95% confidence interval
Smoothed line plot, scatter plot layered
Stacked bar chart
dodged bar chart
stacked KDE Plot

```python
mpg = pd.read_csv('https://raw.githubusercontent.com/tidyverse/ggplot2/master/data-raw/mpg.csv')
diamonds = pd.read_csv('https://raw.githubusercontent.com/tidyverse/ggplot2/master/data-raw/diamonds.csv')
ts = pd.read_csv('random_walk_timeseries.csv')
```

```python
mpg.head()
```

```python
ggplot(mpg) + geom_bar(aes(x='manufacturer')) + coord_flip() + aes(fill='class')
```

```python
ggplot(mpg) + geom_histogram(aes(x='hwy'), bins=12)
```

```python
ggplot(mpg, aes(x='displ', y='hwy', color='class')) + \
geom_point() + \
ggtitle("Highway MPG vs Engine Displacement by Class") + \
xlab("Engine Displacement (L)") + \
ylab("Highway MPG")
```

```python
ggplot(mpg, aes(x='cty', y='hwy', size='cyl')) + \
geom_point(alpha=0.5)
```

```python
ggplot(mpg, aes(x='displ', y='hwy')) \
+ geom_point() \
+ facet_wrap(facets='~class', nrow=2)
```

```python
ggplot(mpg, aes(x='displ', y='hwy')) \
+ geom_point() \
+ facet_grid(facets='drv ~ cyl')
```

```python
ggplot(mpg, aes(x='displ', y='hwy')) + \
geom_point() + \
geom_smooth(method='lm') + \
ggtitle("Highway MPG vs Engine Displacement by Class") + \
xlab("Engine Displacement (L)") + \
ylab("Highway MPG")
```

```python
diamonds.head()
```

```python
ggplot(diamonds, aes(x='cut', fill='clarity')) \
+ geom_bar()
```

```python
ggplot(diamonds, aes(x='cut', fill='clarity')) \
+ geom_bar(position='dodge')
```

```python
ggplot(diamonds, aes(x='depth', color='cut', fill='cut')) \
+ geom_density(aes(y=after_stat('count')), alpha=0.1) \
+ xlim([55, 70])
```
